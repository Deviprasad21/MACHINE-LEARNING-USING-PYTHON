```python
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest, chi2
import pandas as pd

# Load dataset
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target

# Apply SelectKBest with chi2
selector = SelectKBest(score_func=chi2, k=2)
X_new = selector.fit_transform(X, y)

# Show selected features
selected_features = X.columns[selector.get_support()]
print("Selected Features:", selected_features.tolist())
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load dataset
X = iris.data

# Apply PCA to reduce to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the results
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=iris.target)
plt.title("PCA Feature Extraction (Iris Dataset)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Example user-item ratings
data = {'User1': [5, 0, 3, 0],
        'User2': [4, 0, 0, 2],
        'User3': [0, 3, 0, 5],
        'User4': [0, 0, 4, 4]}

df = pd.DataFrame(data, index=['ItemA', 'ItemB', 'ItemC', 'ItemD'])

# User similarity matrix using cosine similarity
user_similarity = cosine_similarity(df.T)
user_similarity_df = pd.DataFrame(user_similarity, index=df.columns, columns=df.columns)

print("User Similarity Matrix:")
print(user_similarity_df)
```
