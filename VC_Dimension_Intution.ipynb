```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
from itertools import product

# Function to test if points can be shattered by a linear classifier
def can_shatter(points):
    for labels in product([0, 1], repeat=len(points)):
        # Skip if only one class present (cannot train classifier with one class)
        if len(set(labels)) < 2:
            continue
        clf = Perceptron(tol=1e-3)
        clf.fit(points, labels)
        preds = clf.predict(points)
        if not np.array_equal(preds, labels):
            return False
    return True

# Test different numbers of points
np.random.seed(42)
for n_points in range(1, 6):
    points = np.random.rand(n_points, 2)
    result = can_shatter(points)
    print(f"{n_points} points can {'be shattered' if result else 'not be shattered'} by a linear classifier.")

# Visualization: plot 3 random points in 2D
points = np.random.rand(3, 2)
plt.scatter(points[:, 0], points[:, 1], c='blue')
plt.title("Example: 3 random points in 2D")
plt.xlabel("X1")
plt.ylabel("X2")
plt.grid(True)
plt.show()
```
