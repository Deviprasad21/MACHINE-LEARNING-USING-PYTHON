```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# XOR input and output
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

y = np.array([0, 1, 1, 0])  # XOR output

# Scaling input features (Optional but recommended)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Different activation functions
activation_functions = ['logistic', 'relu', 'tanh']

# Store results
results = {}

# Train and test the model with different activation functions
for activation in activation_functions:
    model = MLPClassifier(hidden_layer_sizes=(2,), activation=activation, solver='adam', max_iter=10000)
    model.fit(X_scaled, y)

    # Make predictions
    y_pred = model.predict(X_scaled)
    accuracy = accuracy_score(y, y_pred)

    # Save results
    results[activation] = accuracy
    print(f"Activation Function: {activation}")
    print(f"Accuracy: {accuracy:.4f}")
    print("Predictions:", y_pred)
    print("-" * 30)

# Plotting the results
plt.bar(results.keys(), results.values(), color='skyblue')
plt.xlabel('Activation Function')
plt.ylabel('Accuracy')
plt.title('Accuracy of Neural Network on XOR Problem with Different Activation Functions')
plt.show()
```
