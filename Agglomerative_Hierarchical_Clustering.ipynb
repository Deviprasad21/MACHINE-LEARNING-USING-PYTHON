```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate synthetic 2D data
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=42)

# Dendrogram using linkage
linked = linkage(X, method='ward')

plt.figure(figsize=(10, 6))
dendrogram(linked)
plt.title('Dendrogram (Agglomerative Hierarchical Clustering)')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.show()

# Agglomerative clustering with 4 clusters
agg_clust = AgglomerativeClustering(n_clusters=4, linkage='ward')
labels = agg_clust.fit_predict(X)

# Plot the clustered data points
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', s=50)
plt.title('Agglomerative Hierarchical Clustering (4 Clusters)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
